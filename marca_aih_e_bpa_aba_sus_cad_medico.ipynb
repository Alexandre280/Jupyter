```python
nome_caminho_arquivo = "C:\\Users\\6324102\Documents\Python Scripts\Jupyter\lista_usuarios_tasy\\Cadastro Médico Editado-cbo-aih-bpa_13-07-2022.xlsx"
planilha_aberta = load_workbook(filename = nome_caminho_arquivo)

#seleciona a aba que tem as informações a serem passadas para o formulario
sheet_selecionada = planilha_aberta['Planilha1']

pagi.alert('Preenchendo CBO, AHI E OU BPA função medico')

data = datetime.now().strftime('%Y-%m-%d')
hora = datetime.now().strftime('%H:%M:%S')
print('\n\nInicio', 'Data: ', data, '- Hora: ', hora)

#+++++====================++++==============CAMPOS TASY

campo_cpf = '//*[@id="detail_4_container"]/div[2]/div[2]/tasy-wtextbox/div/div/input'
botao_filtrar = '//*[@id="app-view"]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/corpesf7/div/div[4]/div[2]/tasy-wfilter/div/div[2]/div/button'
botão_salvar_sus = '//*[@id="layout"]/div/div/tasy-wdbpanel/div/div[2]/div[4]/div[2]/tasy-wmdetail/div[2]/div[3]/form/w-persistence-action-bar/w-persistence-action-bar-button[1]/tasy-wbutton/div'
botao_funil_filtro = '//*[@id="layout"]/div/div/tasypopupmenu/tasy-wdbpanel/div/div[2]/div[4]/tasy-wlabel'
medico_direita = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/corpesf7/div/div[2]/div[2]/div/div/div/div/tasypopupmenu/tasy-wdbpanel/div/div[2]/div[4]/div[1]/w-datagrid/div[3]/div/div[4]/div[3]/div/div/div[4]/div'
botão_sus = '//*[@id="pages"]/div[5]/ng-include/div/div[2]/div'
botão_ver_sus  = '//*[@id="layout"]/div/div/tasy-wdbpanel/div/div[2]/div[2]/tasy-handlebar-new/div/div[1]/div/button[9]'
campo_cod_cbo  = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/corpesf7/div/div[2]/div[3]/tasy-wtab-panel/div/div[1]/div[2]/tasy-tab-page[5]/tasy-wtab-panel/div/div/div[2]/tasy-tab-page[1]/div/div/div/div/tasy-wdbpanel/div/div[2]/div[4]/div[2]/tasy-wmdetail/div[2]/div[3]/form/div[1]/div[1]/div[2]/tasy-wtextboxlocator/div/div/div[1]/tasy-wtextbox/div/div/input'
check_aih = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/corpesf7/div/div[2]/div[3]/tasy-wtab-panel/div/div[1]/div[2]/tasy-tab-page[5]/tasy-wtab-panel/div/div/div[2]/tasy-tab-page[1]/div/div/div/div/tasy-wdbpanel/div/div[2]/div[4]/div[2]/tasy-wmdetail/div[2]/div[3]/form/div[1]/div[10]/div[2]/div/div/div/input'
botao_aih = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/corpesf7/div/div[2]/div[3]/tasy-wtab-panel/div/div[1]/div[2]/tasy-tab-page[5]/tasy-wtab-panel/div/div/div[2]/tasy-tab-page[1]/div/div/div/div/tasy-wdbpanel/div/div[2]/div[4]/div[2]/tasy-wmdetail/div[2]/div[3]/form/div[1]/div[10]/div[2]/div/div'
check_bpa = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/corpesf7/div/div[2]/div[3]/tasy-wtab-panel/div/div[1]/div[2]/tasy-tab-page[5]/tasy-wtab-panel/div/div/div[2]/tasy-tab-page[1]/div/div/div/div/tasy-wdbpanel/div/div[2]/div[4]/div[2]/tasy-wmdetail/div[2]/div[3]/form/div[1]/div[11]/div[2]/div/div/div/input'
botão_bpa = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/corpesf7/div/div[2]/div[3]/tasy-wtab-panel/div/div[1]/div[2]/tasy-tab-page[5]/tasy-wtab-panel/div/div/div[2]/tasy-tab-page[1]/div/div/div/div/tasy-wdbpanel/div/div[2]/div[4]/div[2]/tasy-wmdetail/div[2]/div[3]/form/div[1]/div[11]/div[2]/div/div'
botão_fechar_sus = '//*[@id="layout"]/div/div/tasy-wdbpanel/div/div[2]/div[2]/tasy-handlebar-new/div/div[1]/div/button[10]'
botão_voltar_medico = '//*[@id="pages"]/div[1]/ng-include/div'
identificar_cbo = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/corpesf7/div/div[2]/div[3]/tasy-wtab-panel/div/div[1]/div[2]/tasy-tab-page[5]/tasy-wtab-panel/div/div/div[2]/tasy-tab-page[1]/div/div/div/div/tasy-wdbpanel/div/div[2]/div[4]/div[1]/w-datagrid/div[3]/div/div[4]/div[3]/div/div/div[1]/div/span'

#+++++================++====+++==============

for linha in range(256, len(sheet_selecionada['A']) +1):
    
#for linha in range(2,3):
    
    #pega a informação dentro da planilha
    nome2 = sheet_selecionada['A%s' % linha].value
    cns2 = sheet_selecionada['B%s' % linha].value
    cpf2 = sheet_selecionada['C%s' % linha].value
    cbo2 = sheet_selecionada['D%s' % linha].value
    amb2 = sheet_selecionada['E%s' % linha].value
    hosp2 = sheet_selecionada['F%s' % linha].value
    
    nome = unidecode(str(nome2))
    cns = str(cns2)
    cpf = str(cpf2)
    cbo = str(cbo2)
    amb = str(amb2)
    hosp = str(hosp2)
    
    #PREENCHE O CAMPO CPF COM CPF DO MEDICO
    time.sleep(1)
    meuNavegador.find_element(By.XPATH, campo_cpf).clear()
    time.sleep(1)
    meuNavegador.find_element(By.XPATH, campo_cpf).send_keys(cpf)
    time.sleep(1)
    
    #CLICA NO BOTÃO FILTRAR
    time.sleep(2)
    meuNavegador.find_element(By.XPATH, botao_filtrar).click()
    time.sleep(1)
    
    #DOUBLE CLICK NO MEDICO DIREITA
    time.sleep(2)
    double_click_meidico_D = meuNavegador.find_element(By.XPATH, medico_direita)
    time.sleep(0.5)
    ActionChains(meuNavegador).double_click(double_click_meidico_D).perform()
    
    #CLICAR BOTÃO SUS
    time.sleep(2)
    meuNavegador.find_element(By.XPATH, botão_sus).click()
    time.sleep(1)
    
    #CLICAR BOTÃO VER (SUS)
    time.sleep(2)
    meuNavegador.find_element(By.XPATH, botão_ver_sus).click()
    time.sleep(1)
    
    
    #CLICAR BOTÃO VER (SUS)
    cbo3 = meuNavegador.find_element(By.XPATH, identificar_cbo).get_attribute("innerText")
    if not cbo3 or cbo3 == '' or cbo3 == 'None':
        time.sleep(1)
        meuNavegador.find_element(By.XPATH, campo_cod_cbo).clear()
        time.sleep(1)
        meuNavegador.find_element(By.XPATH, campo_cod_cbo).send_keys(cbo)
        time.sleep(1)
    
    
    #VERIFICAR SE O CHECKBOX AIH ESTÁ MARCADO
    aih_marcado = ''
    time.sleep(0.5)
    checkbox_aih = meuNavegador.find_element(By.XPATH, check_aih).is_selected()
    time.sleep(0.5)
    if checkbox_aih == False and int(hosp) > 0:
        meuNavegador.find_element(By.XPATH, botao_aih).click()
        aih_marcado = 'AIH marcado pelo robo'
        
    elif checkbox_aih == True:
        aih_marcado = 'AIH ja tava marcado'
        
    else:
        aih_marcado = 'AIH não é pra marcar'
        
    
    #VERIFICAR SE O CHECKBOX BPA ESTÁ MARCADO
    bpa_marcado = ''
    time.sleep(0.5)
    checkbox_bpa = meuNavegador.find_element(By.XPATH, check_bpa).is_selected()
    time.sleep(0.5)
    if checkbox_bpa == False and int(amb) > 0:
        meuNavegador.find_element(By.XPATH, botão_bpa).click()
        bpa_marcado = 'BPA marcado pelo robo'
        
    elif checkbox_bpa == True:
        bpa_marcado = 'BPA ja marcado'
        
    else:
        bpa_marcado = 'BPA não é pra marcar'
        
    if aih_marcado == 'AIH marcado pelo robo' or bpa_marcado == 'BPA marcado pelo robo':
        #CLICA EM SALVAR
        time.sleep(2)
        meuNavegador.find_element(By.XPATH, botão_salvar_sus).click()
        time.sleep(1)  
    
    else:
        #CLICA EM FECHAR
        time.sleep(2)
        meuNavegador.find_element(By.XPATH, botão_fechar_sus).click()
        time.sleep(1)
    
    #CLICA EM VOLTAR PARA VOLTAR PARA OS CAMPOS MEDICO
    time.sleep(2)
    meuNavegador.find_element(By.XPATH, botão_voltar_medico).click()
    time.sleep(1)
    
    #CLICAR NO FUNIL FILTRO
    time.sleep(2)
    meuNavegador.find_element(By.XPATH, botao_funil_filtro).click()
    time.sleep(1)
    
    
    print('linha: {}'.format(linha))
    print('nome: {}'.format(nome))
    print('cns: {}'.format(cns))
    print('cpf: {}'.format(cpf))
    print('cbo: {}'.format(cbo))
    print('AIH marcado ? {}'.format(aih_marcado))
    print('BPA marcado ? {}'.format(bpa_marcado))
    print('___________===============_________________=============__________\n')

data = datetime.now().strftime('%Y-%m-%d')
hora = datetime.now().strftime('%H:%M:%S')
print('\nPreenchendo CBO, AHI E OU BPA função medico\n\nFinalizado', 'Data: ', data, '- Hora: ', hora)
pagi.alert('Finalizado')
```
