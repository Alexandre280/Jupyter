```python
### prenchimento campo Código profissional (CADASTRO DO MEDICO)

from openpyxl import load_workbook
import os
from unidecode import unidecode
import pyautogui as pagi
import time
from datetime import datetime

nome_caminho_arquivo = "C:\\Users\\6324102\Documents\Alexandre\PYTHON AUTOMATIZAÇÃO\Jupyter\lista_usuarios_tasy\\cadastro medico completo.xlsx"
planilha_aberta = load_workbook(filename = nome_caminho_arquivo)

#seleciona a aba que tem as informações a serem passadas para o formulario
sheet_selecionada = planilha_aberta['Planilha1']

pagi.alert('Atualização CEP MEDICO Tasy')
pagi.alert('Lembrete, Desmarcar\n Cadastro médico e Somente médico ativos\n caso não forem medicos\n DA O PLAY\n COLOCAR RESOLUÇÃO DE TELA EM 75%')

data = datetime.now().strftime('%Y-%m-%d')
hora = datetime.now().strftime('%H:%M:%S')
print('Atualização  CEP MEDICO Tasy\n\nInicio', 'Data: ', data, '- Hora: ', hora)

for linha in range(168, len(sheet_selecionada['A']) +1):
    
    matricula2 = sheet_selecionada['A%s' % linha].value
    nome2 = sheet_selecionada['B%s' % linha].value
    cpf2 = sheet_selecionada['C%s' % linha].value
    cep2 = sheet_selecionada['Q%s' % linha].value
    complemento2 = sheet_selecionada['M%s' % linha].value
    numero2 = sheet_selecionada['L%s' % linha].value
    telefone2 = sheet_selecionada['V%s' % linha].value
    
    matricula = str(matricula2)
    nome = unidecode(str.upper(nome2))
    cpf = str(cpf2)
    cep = str(cep2)
    complemento = str(complemento2)
    numero = str(numero2)
    telefone = str(telefone2)
    
    
#=========--------PESQUISA CPF
    #procura campo CPF
    pagi.moveTo(x=54, y=184)
    time.sleep(0.5)
    pagi.click()
    time.sleep(3)
    
    #seleciona cpf no campo
    pagi.moveTo(x=212, y=235)
    pagi.click()
    time.sleep(1.5)
    pagi.write('n')
    time.sleep(1)
    pagi.press('down')
    time.sleep(1)
    pagi.click()
    time.sleep(3)
    
    #campo descrição e coloca cpf
    pagi.moveTo(x=196, y=300)
    pagi.click()
    time.sleep(1)
    pagi.write(cpf)
    
    #filtra a pesquisa
    pagi.moveTo(x=486, y=784)
    time.sleep(1)
    pagi.click()
    
    #seleciona o paciente
    time.sleep(4)
    pagi.moveTo(x=1542, y=788)
    time.sleep(1)
    pagi.click()
    
    # filtra o paciente encontrado / ir para os campos de endereço
    time.sleep(2)
    pagi.moveTo(x=1542, y=788)
    time.sleep(1)
    pagi.click()
    
    #clicar em residencial
    time.sleep(2)
    pagi.moveTo(x=935, y=232)
    time.sleep(1)
    pagi.click()
    
    #adionar cpf
    time.sleep(2)
    pagi.moveTo(x=1542, y=279)
    time.sleep(1)
    pagi.click()
    time.sleep(2)
    
    #coloca o cpf
    pagi.moveTo(x=1341, y=347)
    time.sleep(1)
    pagi.click()
    time.sleep(1)
    pagi.hotkey('ctrlleft', 'a')
    time.sleep(1)
    pagi.press('backspace')
    time.sleep(1)
    pagi.write(cep)
    time.sleep(1)
    pagi.press('tab')
    
    #coloca numero
    time.sleep(1)
    pagi.moveTo(x=1232, y=407)
    time.sleep(1)
    pagi.click()
    pagi.hotkey('ctrlleft', 'a')
    time.sleep(1)
    pagi.press('backspace')
    time.sleep(1)
    pagi.write(numero)
    
    #escrever complemento
    time.sleep(1)
    pagi.moveTo(x=1001, y=522)
    time.sleep(1)
    pagi.click()
    pagi.hotkey('ctrlleft', 'a')
    time.sleep(1)
    pagi.press('backspace')
    time.sleep(1)
    if complemento != 'None':
        pagi.write(complemento)
    
    #descer pagina
    time.sleep(1)
    pagi.moveTo(x=1568, y=657)
    time.sleep(1)
    pagi.click()
    time.sleep(1.5)
    
    #colocar pais
    pagi.moveTo(x=1018, y=426)
    pagi.click()
    time.sleep(6)
    pagi.write('brasil')
    time.sleep(1.5)
    pagi.press('enter')
    
    
    #escrever telefone
    pagi.moveTo(x=985, y=498)
    time.sleep(1)
    pagi.click()
    pagi.hotkey('ctrlleft', 'a')
    time.sleep(1)
    pagi.press('backspace')
    time.sleep(1)
    pagi.write(telefone)
    
    #salvar
    time.sleep(1)
    pagi.moveTo(x=1541, y=786)
    time.sleep(1)
    pagi.click()
    time.sleep(3)
    
    #clica em ok caso haja duplicidade
    pagi.moveTo(x=939, y=607)
    time.sleep(0.5)
    pagi.click()
    time.sleep(3)
    
    #--------------------------------------------------------|=Printar o nome/coren cadastrado do usuario criado=|------------------
    print('linha: ', linha, '\nmatricula: ',matricula)
    print('nome: ', nome)
    print('cpf: ', cpf)
    print('cep: ', cep)
    print('complemento: ', complemento)
    print('numero: ', numero)
    print('telefone: ', telefone)
    print('estabelecimento: OK')
    print('___________===============_________________=============__________\n')

data = datetime.now().strftime('%Y-%m-%d')
hora = datetime.now().strftime('%H:%M:%S')
print('\nAtualização CEP MEDICO Tasy\n\nFinalizado', 'Data: ', data, '- Hora: ', hora)
pagi.alert('Finalizado')
    
    
```
