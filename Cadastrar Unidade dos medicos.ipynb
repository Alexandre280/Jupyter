```python
# prenchimento campo Código profissional (CADASTRO DO MEDICO)

from openpyxl import load_workbook
import os
from unidecode import unidecode
import pyautogui as pagi
import time
from datetime import datetime

nome_caminho_arquivo = "C:\\Users\\6324102\Documents\Alexandre\PYTHON AUTOMATIZAÇÃO\Jupyter\lista_usuarios_tasy\\cadastro medico completo.xlsx"
planilha_aberta = load_workbook(filename = nome_caminho_arquivo)

#seleciona a aba que tem as informações a serem passadas para o formulario
sheet_selecionada = planilha_aberta['Planilha1']

pagi.alert('Atualização CEP MEDICO Tasy')
pagi.alert('Lembrete, Desmarcar\n Cadastro médico e Somente médico ativos\n caso não forem medicos\n DA O PLAY\n COLOCAR RESOLUÇÃO DE TELA EM 75%')

data = datetime.now().strftime('%Y-%m-%d')
hora = datetime.now().strftime('%H:%M:%S')
print('Atualização  CEP MEDICO Tasy\n\nInicio', 'Data: ', data, '- Hora: ', hora)

for linha in range(2, len(sheet_selecionada['A']) +1):
    
    matricula2 = sheet_selecionada['A%s' % linha].value
    nome2 = sheet_selecionada['B%s' % linha].value
    cpf2 = sheet_selecionada['C%s' % linha].value
    cep2 = sheet_selecionada['Q%s' % linha].value
    complemento2 = sheet_selecionada['M%s' % linha].value
    numero2 = sheet_selecionada['L%s' % linha].value
    
    matricula = str(matricula2)
    nome = unidecode(str.upper(nome2))
    cpf = str(cpf2)
    cep = str(cep2)
    complemento = str(complemento2)
    numero = str(numero2)
    
    
    #=========--------PESQUISA CPF
    #procura campo CPF
    pagi.moveTo(x=747, y=317)
    time.sleep(0.5)
    pagi.click()
    time.sleep(0.5)

    #digita o cpf
    pagi.write(cpf)
    time.sleep(0.5)
    
#filtrar busca
    pagi.moveTo(x=988, y=774)
    time.sleep(0.5)
    pagi.click()
    
    time.sleep(3)
    
#ver o medico
    pagi.moveTo(x=1134, y=284)
    time.sleep(0.5)
    pagi.doubleClick()
    time.sleep(1.5)
    
#Relacionamento
    pagi.moveTo(x=209, y=187)
    time.sleep(0.5)
    pagi.click()
    time.sleep(1)

#Estabelecimento
    pagi.moveTo(x=60, y=740)
    time.sleep(0.5)
    pagi.click()
    time.sleep(1)

#adcionar Estabelecimento
    pagi.moveTo(x=913, y=542)
    time.sleep(0.5)
    pagi.click()
    time.sleep(1)
    pagi.hotkey('ctrlleft', 'altleft', 'v')

#Mudar Estabelecimento
    pagi.moveTo(x=614, y=344)
    time.sleep(0.5)
    pagi.click()
    time.sleep(1)
    pagi.write('hmmu')
    time.sleep(1)
    pagi.press('enter')
    time.sleep(1)

#salvar Estabelecimento
    pagi.moveTo(x=1534, y=786)
    time.sleep(0.5)
    pagi.click()
    time.sleep(2.5)

#voltar para filtro67211895691
    pagi.moveTo(x=100, y=158)
    time.sleep(0.5)
    pagi.click()
    time.sleep(2)
    
#clicar no filtro
    pagi.moveTo(x=34, y=260)
    time.sleep(0.5)
    pagi.click()
    time.sleep(0.5)   

#--------------------------------------------------------|=Printar o nome/coren cadastrado do usuario criado=|------------------
    print('linha: ', linha, '\nmatricula: ',matricula)
    print('nome: ', nome)
    print('cpf: ', cpf)
    print('estabelecimento: OK')
    print('___________===============_________________=============__________\n')

data = datetime.now().strftime('%Y-%m-%d')
hora = datetime.now().strftime('%H:%M:%S')
print('\nAtualização COREN Tasy\n\nFinalizado', 'Data: ', data, '- Hora: ', hora)
pagi.alert('Finalizado')
```
