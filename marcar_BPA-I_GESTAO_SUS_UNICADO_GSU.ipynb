```python
from selenium import webdriver as opções_selenium_aula
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pyautogui as pagi
from openpyxl import load_workbook
import os
from unidecode import unidecode
import time
from datetime import datetime
from selenium.webdriver import ActionChains

meuNavegador = opções_selenium_aula.Chrome()
#meuNavegador.get("www.tasyweb.spdmafiliadas.org.br")
```

# EXECUTA PROCEDIMENTO NO TASY

```python
nome_caminho_arquivo = "C:\\Users\\6324102\Documents\Python Scripts\Jupyter\lista_usuarios_tasy\\cadastros_BPA_I_FATURAMENTO.xlsx"
planilha_aberta = load_workbook(filename = nome_caminho_arquivo)

#seleciona a aba que tem as informações a serem passadas para o formulario
sheet_selecionada = planilha_aberta['Planilha1']

pagi.alert('MARCAR marcar BPA-I GESTAO SUS UNICADO GSU função GESTAO SUS UNICADO GSU')

data = datetime.now().strftime('%Y-%m-%d')
hora = datetime.now().strftime('%H:%M:%S')
print('\n\nInicio', 'Data: ', data, '- Hora: ', hora)

#+++++====================++++==============CAMPOS TASY

funil_filtro = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/atesussu/div/div[2]/div[1]/div/div/div/div/tasypopupmenu/tasy-wdbpanel/div/div[2]/div[2]/div[1]/tasy-wlabel'
campo_cod_proced = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/atesussu/div/div[4]/div[2]/tasy-wfilter/div/tasy-wmdetail/div[2]/div[3]/form/div[1]/div[1]/div[2]/tasy-wtextboxlocator/div/div/div[1]/tasy-wtextbox/div/div/input'
botao_filtra_proc = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/atesussu/div/div[4]/div[2]/tasy-wfilter/div/div[2]/div/button'
abre_aba_D = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/atesussu/div/div[2]/div[2]/div[2]/div[1]/tasy-listbox/div'
abre_aba_D_A_B_A = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/atesussu/div/div[2]/div[2]/div[2]/div[1]/tasy-listbox/div/div/a/ng-include/span'
botao_registro_BPA = '/html/body/div[4]/div[2]/div[1]/div/a[14]/span'
botao_adicionar_C = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/atesussu/div/div[2]/div[2]/div[2]/div[2]/div/div/div/div/tasy-wdbpanel/div/div[2]/div[4]/div[1]/w-datagrid/div[1]/span/div/button'
botao_adicionar_D  = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/atesussu/div/div[2]/div[2]/div[2]/div[2]/div/div/div/div/tasy-wdbpanel/div/div[2]/div[2]/tasy-handlebar-new/div/div[1]/div/button[1]'
botao_BPAI  = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/atesussu/div/div[2]/div[2]/div[2]/div[2]/div/div/div/div/tasy-wdbpanel/div/div[2]/div[4]/div[2]/tasy-wmdetail/div[2]/div[3]/form/div[1]/div[1]/div[2]/tasy-wradiobuttongroup/div/div/div[2]/span/div/label/div'
botao_salvar_D = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/atesussu/div/div[2]/div[2]/div[2]/div[2]/div/div/div/div/tasy-wdbpanel/div/div[2]/div[4]/div[2]/tasy-wmdetail/div[2]/div[3]/form/w-persistence-action-bar/w-persistence-action-bar-button[1]/tasy-wbutton/div'
botao_fechar_E = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/atesussu/div/div[2]/div[1]/div/div/div/div/tasypopupmenu/tasy-wdbpanel/div/div[2]/div[2]/tasy-handlebar-new/div/div[1]/div/button[10]'
funil_filtro_final = '/html/body/div[1]/tasy-corsisf1/div/w-mainlayout/div/div/feature-instance/atesussu/div/div[2]/div[1]/div/div/div/div/tasypopupmenu/tasy-wdbpanel/div/div[2]/div[4]/tasy-wlabel'

#+++++================++====+++==============

filtro = 0

for linha in range(518, len(sheet_selecionada['A']) +1):
    
#for linha in range(2,3):
    
    #pega a informação dentro da planilha
    cod_BPA2 = sheet_selecionada['A%s' % linha].value
    nome_proced2 = sheet_selecionada['B%s' % linha].value
    
    cod_BPA = unidecode(str(cod_BPA2))
    nome_proced = str(nome_proced2)
    
    if filtro == 0:
        #CLICA NO BOTÃO FUNIL PARA ABRIR O FILTRO
        time.sleep(2)
        meuNavegador.find_element(By.XPATH, funil_filtro).click()
        time.sleep(1)
        filtro = 1
        
    else:
        #CLICA NO BOTÃO FUNIL PARA ABRIR O FILTRO
        time.sleep(2)
        meuNavegador.find_element(By.XPATH, funil_filtro_final).click()
        time.sleep(1)
    
    #PREENCHE O CAMPO CODIGO COM O CODIGO DO PROCEDIMENTO
    time.sleep(1)
    meuNavegador.find_element(By.XPATH, campo_cod_proced).clear()
    time.sleep(1)
    meuNavegador.find_element(By.XPATH, campo_cod_proced).send_keys(cod_BPA)
    time.sleep(1)

    #CLICA NO BOTÃO FILTRAR
    time.sleep(1)
    meuNavegador.find_element(By.XPATH, botao_filtra_proc).click()
    time.sleep(1)

    if meuNavegador.find_element(By.XPATH, abre_aba_D).is_displayed():
        #CLICA NA ABA PARA ABRIR MAIS OPÇÕES SEM OPÇÕES
        time.sleep(1)
        meuNavegador.find_element(By.XPATH, abre_aba_D).click()
        time.sleep(1)
    else:
        #CLICA NA ABA PARA ABRIR MAIS OPÇÕES COM A OPÇÃO "APAC/BPA/AIH"
        time.sleep(1)
        meuNavegador.find_element(By.XPATH, abre_aba_D_A_B_A).click()
        time.sleep(1)

    #CLICA NA OPÇÃO DA ABA Registro BPA
    time.sleep(1)
    meuNavegador.find_element(By.XPATH, botao_registro_BPA).click()
    time.sleep(1)

    #VERIFICA SE NÃO TEM REGISTRO JA CADASTRADO
    if meuNavegador.find_element(By.XPATH, botao_adicionar_C).is_displayed():
        meuNavegador.find_element(By.XPATH, botao_adicionar_C).click()
        time.sleep(1)
    else:
        #CLICA EM ADICIONAR PARA REGISTRAR O BPA
        time.sleep(1)
        meuNavegador.find_element(By.XPATH, botao_adicionar_D).click()
        time.sleep(1)

    #MARCA O CHECKBOX O BOTÃO BPA I
    time.sleep(1)
    meuNavegador.find_element(By.XPATH, botao_BPAI).click()
    time.sleep(1)

    #SALVA O REGISTRO CADASTRADO
    time.sleep(1)
    meuNavegador.find_element(By.XPATH, botao_salvar_D).click()
    time.sleep(1)

    #FECHAR PROCEDIMENTO
    time.sleep(1)
    meuNavegador.find_element(By.XPATH, botao_fechar_E).click()
    time.sleep(1)
    
    print('linha: {}'.format(linha))
    print('código procedimento: {}'.format(cod_BPA))
    print('procedimento: {}'.format(nome_proced))
    print('___________===============_________________=============__________\n')

data = datetime.now().strftime('%Y-%m-%d')
hora = datetime.now().strftime('%H:%M:%S')
print('\nMARCAR marcar BPA-I GESTAO SUS UNICADO GSU função GESTAO SUS UNICADO GSU\n', 'Data: ', data, '- Hora: ', hora)
pagi.alert('Finalizado')
```
