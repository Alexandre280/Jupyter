```python
##### prenchimento campo Código profissional (CADASTRO DO MEDICO)

from openpyxl import load_workbook
import os
from unidecode import unidecode
import pyautogui as pagi
import time
from datetime import datetime

nome_caminho_arquivo = "C:\\Users\\6324102\Documents\Alexandre\PYTHON AUTOMATIZAÇÃO\Jupyter\lista_usuarios_tasy\\assistencial atualizado para cadatro 20-04-2022.xlsx"
planilha_aberta = load_workbook(filename = nome_caminho_arquivo)

#seleciona a aba que tem as informações a serem passadas para o formulario
sheet_selecionada = planilha_aberta['Planilha1']

pagi.alert('Atualização COREN Tasy')
pagi.alert('Lembrete, Desmarcar\n Cadastro médico e Somente médico ativos\n caso não forem medicos\n DA O PLAY\n COLOCAR RESOLUÇÃO DE TELA EM 75%')

data = datetime.now().strftime('%Y-%m-%d')
hora = datetime.now().strftime('%H:%M:%S')
print('Atualização COREN Tasy\n\nInicio', 'Data: ', data, '- Hora: ', hora)

for linha in range(546, len(sheet_selecionada['A']) +1):

    #pega a informação dentro da planilha
    matricula2 = sheet_selecionada['A%s' % linha].value
    nome2 = sheet_selecionada['B%s' % linha].value
    cpf2 = sheet_selecionada['C%s' % linha].value
    coren2 = sheet_selecionada['D%s' % linha].value
    tipo_C2 = sheet_selecionada['E%s' % linha].value
    funcao2 = sheet_selecionada['F%s' % linha].value
    dtN2 = sheet_selecionada['G%s' % linha].value
    sexo2 = sheet_selecionada['H%s' % linha].value
    cns2 = sheet_selecionada['I%s' % linha].value
    #admissao2 = sheet_selecionada['J%s' % linha].value
    #endereco2 = sheet_selecionada['K%s' % linha].value
    #numero2 = sheet_selecionada['L%s' % linha].value
    #complemento2 = sheet_selecionada['M%s' % linha].value
    #bairro2 = sheet_selecionada['N%s' % linha].value
    #cidade2 = sheet_selecionada['O%s' % linha].value
    #uf2 = sheet_selecionada['P%s' % linha].value
    #cep2 = sheet_selecionada['Q%s' % linha].value
    #grau_escola2 = sheet_selecionada['R%s' % linha].value
    cor2 = sheet_selecionada['S%s' % linha].value
    #nacionalidade2 = sheet_selecionada['T%s' % linha].value
    estado_civil2 = sheet_selecionada['U%s' % linha].value
    #telefone2 = sheet_selecionada['V%s' % linha].value
    
    #coloca as informações da planilhas nas variaveis
    matricula = str(matricula2)
    nome = unidecode(str.upper(nome2))
    cpf = str(cpf2)
    
    '''if len(cpf1) > 11:
        cpf = cpf1[1:]
    else:
        cpf = cpf1'''
    
    coren = str(coren2)
    tipo_C = str.lower(tipo_C2)
    funcao = str.lower(funcao2)
    dtN = str(dtN2.strftime('%d-%m-%Y'))
    sexo = str.lower(sexo2)
    cns = str(cns2)
    #admissao = str(admissao2.strftime('%d-%m-%Y'))
    #endereco = str.lower(endereco2)
    #numero = str(numero2)
    #complemento = str(complemento2)
    #bairro = str.lower(bairro2)
    #cidade = str.lower(cidade2)
    #uf = str.lower(uf2)
    #cep = str(cep2)
    #grau_escola = str(grau_escola2)
    cor = str.lower(cor2)
    #nacionalidade = str.lower(nacionalidade2)
    estado_civil = str.lower(estado_civil2)
    #telefone = str(telefone2)

    
    
    
    #=========--------PESQUISA CPF
    #procura campo CPF
    pagi.moveTo(x=747, y=317)
    time.sleep(0.5)
    pagi.click()
    time.sleep(0.5)

    #digita o cpf
    pagi.write(cpf)
    time.sleep(0.5)
    
#filtrar busca
    pagi.moveTo(x=988, y=774)
    time.sleep(0.5)
    pagi.click()
    
    time.sleep(3)
    
#adicionar pessoa
    pagi.moveTo(x=408, y=556)
    time.sleep(0.5)
    pagi.click()
    
#ver o medico
    pagi.moveTo(x=675, y=206)
    time.sleep(0.5)
    #pagi.click()
    pagi.hotkey('ctrlleft', 'altleft', 'v')
    time.sleep(0.5)

#escreve a matricula do funcionario
    pagi.moveTo(x=372, y=468)
    time.sleep(0.5)
    pagi.click()
    time.sleep(0.5)
    pagi.hotkey('ctrlleft', 'a')
    time.sleep(0.5)
    pagi.press('backspace')
    time.sleep(0.5)
    pagi.write(matricula)
    
    
#seleciona o tipo de conselho
    pagi.moveTo(x=701, y=460)
    time.sleep(0.5)
    pagi.click()
    time.sleep(1)
    pagi.click()
    pagi.write(tipo_C)
    time.sleep(1)
    
#digita codigo profissional referente ao conselho
    pagi.moveTo(x=211, y=528)
    time.sleep(0.5)
    pagi.click()
    pagi.write(coren)
    time.sleep(0.5)
    
#digitar Profissão/Ocupação (função)
    pagi.moveTo(x=553, y=344)
    time.sleep(0.5)
    pagi.click()
    time.sleep(0.5)
    
    if funcao == 'tecnico de enfermagem': #se for tecnico de enfermagem
        pagi.write(funcao)
        time.sleep(1)
        pagi.press('down')
        time.sleep(1)
        pagi.press('enter')
        
    else: #se qualquer outro tipo de função
        pagi.write(funcao)
        time.sleep(1)
        pagi.press('enter')
    
#digita nome 
    pagi.moveTo(x=388, y=293) #clica no campo nome
    pagi.click()
    pagi.hotkey('ctrlleft', 'a')
    time.sleep(0.5)
    pagi.press('backspace')
    time.sleep(1.0)
    pagi.write(nome)
    time.sleep(1.0)
    
#digita data nascimento
    pagi.moveTo(x=175, y=758)
    pagi.click()
    pagi.hotkey('ctrlleft', 'a')
    time.sleep(0.5)
    pagi.press('backspace')
    pagi.write(dtN)
    
#digita cns
    if cns != '#N/A':
        pagi.moveTo(x=195, y=703)
        pagi.click()
        pagi.hotkey('ctrlleft', 'a')
        time.sleep(0.5)
        pagi.press('backspace')
        pagi.write(cns)
    
#seleciona sexo
    pagi.moveTo(x=695, y=286)
    time.sleep(0.5)
    pagi.click()
    time.sleep(0.5)
    if sexo == 'masculino':
        pagi.press('m')
        time.sleep(0.5)
        pagi.press('enter')
        
    else:
        pagi.press('f')
        time.sleep(0.5)
        pagi.press('enter')
    time.sleep(0.5)

    
#selecionar nacionalidade
    pagi.moveTo(x=531, y=522)
    time.sleep(0.5)
    pagi.click()
    time.sleep(0.5)
    pagi.write('bra')
    time.sleep(0.5)
    pagi.press('enter')
    
#estado civil
    pagi.moveTo(x=151, y=346)
    time.sleep(0.5)
    pagi.click()
    time.sleep(0.5)
    if estado_civil == 'casado':
        pagi.write('casa')
        time.sleep(0.5)
        pagi.press('enter')
        time.sleep(0.5)

    elif estado_civil == 'solteiro':
        pagi.write('sol')
        time.sleep(0.5)
        pagi.press('enter')
        time.sleep(0.5)

    elif estado_civil == 'divorciado':
        pagi.write('div')
        time.sleep(0.5)
        pagi.press('enter')
        time.sleep(0.5)

    elif estado_civil == 'união estável':
        pagi.write('con')
        time.sleep(0.5)
        pagi.press('enter')
        time.sleep(0.5)

#selecionar Cor
    pagi.moveTo(x=334, y=346)
    time.sleep(0.5)
    pagi.click()
    time.sleep(0.5)
    if cor == 'branca':
        pagi.write('bra')
        time.sleep(1)
        pagi.press('enter')
        time.sleep(1)

    elif cor == 'amarela':
        pagi.write('ama')
        time.sleep(1)
        pagi.press('enter')
        time.sleep(1)

    elif cor == 'parda':
        pagi.write('par')
        time.sleep(1)
        pagi.press('enter')
        time.sleep(1)

    elif cor == 'preta':
        pagi.write('neg')
        time.sleep(1)
        pagi.press('enter')
        time.sleep(1)

#colocar CPF 
    pagi.moveTo(x=336, y=574)
    pagi.click()
    pagi.hotkey('ctrlleft', 'a')
    time.sleep(0.5)
    pagi.press('backspace')
    time.sleep(1.0)
    pagi.write(cpf)
    time.sleep(1.0)
        
#colocar CBO
    pagi.moveTo(x=773, y=716) #DESCE SCROLL
    time.sleep(1)
    pagi.click()
    time.sleep(1)

    if funcao == 'tecnico de enfermagem': #se for tecnico de enfermagem
        pagi.moveTo(x=461, y=714) #CAMPO CBO
        time.sleep(0.5)
        pagi.click()
        time.sleep(0.5)
        pagi.write(funcao)
        time.sleep(1)
        pagi.press('down')
        time.sleep(1)
        pagi.press('enter')
        
    elif funcao == 'enfermeiro': #se for tecnico de radiologia
        pagi.moveTo(x=461, y=714) #CAMPO CBO
        time.sleep(0.5)
        pagi.click()
        time.sleep(0.5)
        pagi.write(funcao)
        time.sleep(1)
        pagi.press('down')
        time.sleep(1)
        pagi.press('enter')

#salvar dados do medico
    pagi.moveTo(x=740, y=792)
    time.sleep(0.5)
    pagi.click()
    time.sleep(4)
    
#clica em ok caso haja duplicidade
    pagi.moveTo(x=939, y=607)
    time.sleep(0.5)
    pagi.click()
    time.sleep(3)
    
#cancelar lado direito
    pagi.moveTo(x=1462, y=792)
    time.sleep(1.5)
    pagi.click()
    time.sleep(1)
    
#sair sem salvar
    pagi.moveTo(x=694, y=553)
    time.sleep(0.5)
    pagi.click()
    time.sleep(0.5)
    
#fechar cadastro
    pagi.moveTo(x=678, y=220)
    time.sleep(0.5)
    pagi.click()
    time.sleep(0.5)
    
#clicar no filtro
    pagi.moveTo(x=34, y=260)
    time.sleep(0.5)
    pagi.click()
    time.sleep(0.5)   

#--------------------------------------------------------|=Printar o nome/coren cadastrado do usuario criado=|------------------
    print('linha: ', linha, '\nmatricula: ',matricula)
    print('nome: ', nome)
    print('cpf: ', cpf)
    print('conselho: ', coren )
    print('tipo conselho: ', tipo_C)
    print('data nascimento: ', dtN)
    print('cor: ', cor)
    print('estado civil: ', estado_civil)
    print('CNS: ', cns)
    print('___________===============_________________=============__________\n')

data = datetime.now().strftime('%Y-%m-%d')
hora = datetime.now().strftime('%H:%M:%S')
print('\nAtualização COREN Tasy\n\nFinalizado', 'Data: ', data, '- Hora: ', hora)
pagi.alert('Fifnalizado')
```
